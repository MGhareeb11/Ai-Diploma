Write a Python function that takes a list of words and return the longest word and the length of the longest one
(Longest word: Exercises
Length of the longest word: 9)

````python
def longest_word(words):
    longest = words[0]

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest, len(longest)


words = input("Enter words separated by spaces: ").split()

word, length = longest_word(words)

print("Longest word:", word)
print("Length of the longest word:", length)
````

A prime number is a number that is greater than 1 and has only two factors which are 1 and itself. The first few prime numbers are 2, 3, 5, 7, 11, 13, 17, 19, 23 and 29.

Given a number N. Determine whether N is prime or not.

Note: Solve this problem using function.

Input
First line will contain a number T (1â€‰â‰¤â€‰Tâ€‰â‰¤â€‰103) number of test cases.

Next T lines will contain a number N (1â€‰â‰¤â€‰Nâ€‰â‰¤â€‰109).

Output
Print "YES" if the Nth number is prime otherwise, print "NO".

Example

Input

3

2

4

8

Output

YES

NO

NO

````python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2

    return True

t = int(input())

for _ in range(t):
    n = int(input())

    if is_prime(n):
        print("YES")
    else:
        print("NO")
````

Write a function called fibonacci(n) that:

Takes an integer N (1 â‰¤ N â‰¤ 45) as a parameter.

Returns a list containing the first N numbers of the Fibonacci sequence.

Use the following definition:

fib(1) = 0

fib(2) = 1

fib(n) = fib(n - 1) + fib(n - 2)

After defining the function, read an integer N from the user and print the returned Fibonacci sequence numbers separated by spaces.

ðŸ”¹ Example

Input

7

Output

0 1 1 2 3 5 8

Note
For more information visit Fibonacci: https://www.mathsisfun.com/numbers/fibonacci-sequence.html.

````python
def fibonacci(n):
    if n == 1:
        return [0]
    if n == 2:
        return [0, 1]

    fib_list = [0, 1]

    for i in range(2, n):
        next_number = fib_list[i - 1] + fib_list[i - 2]
        fib_list.append(next_number)

    return fib_list


n = int(input())

result = fibonacci(n)

print(*result)
````

````python
def print_pyramid(n, current=1):
    if current > n:
        return

    spaces = n - current
    stars = 2 * current - 1

    print(" " * spaces + "*" * stars)

    print_pyramid(n, current + 1)


n = int(input())

print_pyramid(n)
````

 Write a Python function is_divisible(num1, num2) that takes two numbers as input and returns True if the first number is divisible by the second number, and False otherwise.

````python
def is_divisible(num1, num2):
    if num2 == 0:
        return False
    return num1 % num2 == 0

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print(is_divisible(a, b))
````

Write a function that takes an integer number as input and returns a list containing each digit of the number separately.

````python
def digits_list(number):
    number = abs(number)
    digits = []

    for digit in str(number):
        digits.append(int(digit))

    return digits
n = int(input("Enter a number: "))

result = digits_list(n)

print(result)
````
