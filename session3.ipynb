# Triangle Star Pattern

Input: 4
Output:
* ## *
* ## **
* ## ***
* ## ****


```python
for i in range(5):
    for x in range(i):
        print("*", end=" ")
    print()
```

# Count Vowels in a String
## Take a string input and count how many vowels it contains.

```python
s=input("Enter you String:")
vowels=["a","u","i","e","o"]
count=0
for i in s:
    if i in vowels:
        count+=1
print("vowels:",count)        
```

Given a list of  numbers. Count how many of these values are even, odd, positive and negative.

l=-5 0 -3 -4 12

Output:

Even: 3

Odd: 2

Positive: 1

Negative: 3


```python
l=[-5,0,-3,-4,12]
even=0
odd=0
positive=0
negative=0
for x in l:
    if x %2==0:
        even+=1
    else:
        odd+=1
    if x >0:
        positive+=1
    elif x<0:
        negative+=1
print("positive:",positive)
print("negative:",negative)
print("even:",even)
print("odd:",odd)   
```
